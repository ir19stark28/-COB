```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\galiy\COB\TASK2\dataset - netflix1.csv")

df.head()
```

VISULALIZATIONS


```python
sns.countplot(x='type',data=df,palette='Set2')
plt.title("CONTENT COUNT")
```

```python
plt.figure(figsize=(8,8))
df['type'].value_counts().plot.pie(autopct='%1.1f%%',startangle=90)
plt.title("PIE ~ DIST OF TV SHOWS AND MOVIES")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['release_year'],bins=30,kde=True)
plt.title("Movie Count vs Release Year")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='country', order=df['country'].value_counts().index[0:10],data=df,palette='Set1')
plt.title("Content per Country")
```

```python

counts_data = df.groupby(['release_year', 'type']).size().reset_index(name='count')


plt.figure(figsize=(12, 6))


sns.barplot(x='release_year', y='count', hue='type', order=df['release_year'].value_counts().index[0:20], data=counts_data, palette='Set2')


plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Content Type per year')
plt.legend(title='Type')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='release_year',order=df['release_year'].value_counts().index[0:20],data=df,palette='Set1')
plt.title("Amount of Content on Netflix")
```
