# Cramér's V Calculation for Label-Encoded Features and 'release_year'

The features which are possible to be labelled will be lablled using label encoder.

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()

# labelling type feature
encode.fit(data.type)
new_type=encode.transform(data.type)
data.type= new_type

# labelling country feature
encode.fit(data.country)
new_country=encode.transform(data.country)
data.country= new_country

#labelling date feature
encode.fit(data.date_added)
new_date=encode.transform(data.date_added)
data.date_added= new_date

#labelling rate feature
encode.fit(data.rating)
new_rate=encode.transform(data.rating)
data.rating= new_rate

#labelling rate feature
encode.fit(data.duration)
new_dur=encode.transform(data.duration)
data.duration= new_dur

#labelling show_id feature
encode.fit(data.show_id)
new_id=encode.transform(data.show_id)
data.show_id=new_id



encoded_data = data.copy()


```

```python
from scipy.stats import chi2_contingency


# Initialize a new DataFrame to store Cramér's V values
cramer_vs_df = pd.DataFrame(index=['release_year'], columns=['type', 'country', 'date_added', 'rating', 'duration', 'show_id'])

# Calculate Cramér's V for each label-encoded feature and 'release_year'
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# Calculate Cramér's V for each label-encoded feature and 'release_year'
for col in ['type', 'country', 'date_added', 'rating', 'duration', 'show_id']:
    cramer_vs_df.at['release_year', col] = cramers_v(encoded_data[col], encoded_data['release_year'])

# Display the DataFrame with Cramér's V values
print(cramer_vs_df)
```
